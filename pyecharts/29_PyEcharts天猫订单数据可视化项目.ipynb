### PyEcharts数据可视化

```python
data.head()
```

#### 各省份的订单量

```python
data.收货地址.value_counts()
```

```python
def province_map(p):
    if p in ['北京', '天津', '上海', '重庆']:
        return p + '市'
    return p

data.收货地址 = data.收货地址.map(province_map)
```

```python
data.收货地址.value_counts()
```

```python
# 查看订单付款时间不为空的
data.订单付款时间.notnull()

# 过滤掉（删掉）没有付款的订单数据
data[data.订单付款时间.notnull()]
```

```python
# 统计各个省份的订单量

result = data[data.订单付款时间.notnull()].groupby('收货地址')[['订单编号']].count()
result
```

```python
result2 = result.to_dict()['订单编号']
result2
```

```python
result2.items()
```

```python
list(result2.items())
```

```python
[*result2.items()]
```

```python
c = (
    Map()
    .add('各省份订单量', 
        [*result2.items()], 
         'china',
         is_map_symbol_show=False
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(
        title_opts=opts.TitleOpts(title='各省份订单量'),
        visualmap_opts=opts.VisualMapOpts(max_=2000)
    )
)
c.render_notebook()
```

#### 时间序列分析

```python
data.head()
```

```python
data.info()

# 订单创建时间和订单付款时间 先修改成时间类型
```

```python
data.订单创建时间 = pd.to_datetime(data.订单创建时间)
data.订单付款时间 = pd.to_datetime(data.订单付款时间)
```

```python
data.info()
```

#### 每天订单量统计可视化

```python
data.订单创建时间
```

```python
order_add_time = data.订单创建时间.map(lambda x: x.strftime('%Y-%m-%d'))
```

```python
order_add_time
```

```python
# 按照订单创建时间 分组，得到每一天的订单量
result3 = data.groupby(order_add_time).agg({'订单编号': 'count'}).to_dict()['订单编号']
result3
```

```python
list(result3.keys())
```

```python
c = (
    Line()
    .add_xaxis(list(result3.keys()))
    .add_yaxis(
        '订单量',
        list(result3.values())
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_='max')
            ]
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='每天订单量')
    )
)
c.render_notebook()
```

#### 每小时订单量统计可视化

```python
order_add_time2 = data.订单创建时间.map(lambda x: x.strftime('%H'))
```

```python
order_add_time2.value_counts()
```

```python
result4 = data.groupby(order_add_time2).agg({'订单编号': 'count'}).to_dict()['订单编号']
result4
```

```python
c = (
    Bar()
    .add_xaxis( list(result4.keys()) )
    .add_yaxis('订单量', list(result4.values()))
    
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_='max')
            ]
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='每小时订单量')
    )
)
c.render_notebook()
```
