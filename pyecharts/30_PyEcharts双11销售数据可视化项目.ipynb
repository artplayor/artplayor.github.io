### PyEcharts可视化

#### 每天销售量统计分析

```python
data.head(3)
```

```python
result = data.groupby('update_time').agg({'sale_count': 'sum'}).to_dict()['sale_count']
result
```

```python
c = (
    Line()
    .add_xaxis( list(result.keys()) )
    .add_yaxis('销量', list(result.values()))
    
    .set_global_opts(
        title_opts=opts.TitleOpts(title='每天的销售量')
    )
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_='max', name='最大值'),
                opts.MarkPointItem(type_='min', name='最小值'),
            ]
        )
    )
    
)
c.render_notebook()
```

#### 时间升序

```python
times = list( data.update_time.unique() )
times.reverse()
times
```

```python
data.head(2)
```

```python
data.店名.unique()
```

```python
# 时间轮播图

tl = Timeline()

tl.add_schema(
    is_auto_play=True,  # 是否自动播放
    is_loop_play=True,  # 是否循环播放
    play_interval=2000  # 每隔2秒动一次
)

for dt in times:
    item = data[data.update_time <= dt].groupby('店名') \
            .agg({'sale_count': 'sum', 'sale_amount': 'sum'}) \
            .sort_values(by='sale_count', ascending=False)[:10]  \
            .sort_values(by='sale_count')  \
            .to_dict()
    # print(item)
    
    # 柱形图
    bar = (
        Bar()
        .add_xaxis(list(item['sale_count'].keys()))
        .add_yaxis('销售量',
                  [round(val/1000000,1) for val in item['sale_count'].values()],
                   label_opts=opts.LabelOpts(position='right', 
                                            formatter='{@[1]/}百万件'
                                            )
        )
         .add_yaxis('销售额',
                  [round(val/100000000,1) for val in item['sale_amount'].values()],
                   label_opts=opts.LabelOpts(position='right', 
                                            formatter='{@[1]/}亿元'
                                            )
        )
        
        .reversal_axis()  # 条形图
        .set_global_opts(
            title_opts=opts.TitleOpts(title='累计销售量Top10')
        )
    )
    tl.add(bar, dt)
```

```python
item = data.groupby('店名') \
            .agg({'sale_count': 'sum'}) \
            .sort_values(by='sale_count', ascending=False)[:10]  \
            .to_dict()['sale_count']

item = {k: round(v/10000, 1)  for k, v in item.items()}
item
# 单位是：万
```

```python
c = (
    Pie()
    .add(
        '销售量',
        list(item.items())
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(formatter="{b}: {c}w({d})%")
    )
)
c.render_notebook()
```

#### 价格排序

```python
data.head()
```

```python
item = data.groupby('店名') \
            .agg({'price': 'mean'}) \
            .sort_values(by='price', ascending=True)[:20]  \
            .to_dict()['price']

item
```

```python
c = (
    Bar()
    .add_xaxis(list(item.keys()))
    .add_yaxis(
        '价格',
        [round(v, 2)  for v in item.values()],
        label_opts=opts.LabelOpts(position='right')
    )
    .reversal_axis()
    
    .set_global_opts(title_opts=opts.TitleOpts(title='平均价格Top20'))
)
c.render_notebook()
```
