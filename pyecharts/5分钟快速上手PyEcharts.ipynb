## 5分钟上手PyEcharts

首先开始来绘制你的第一个图表

```python
from pyecharts.charts import Bar

bar = Bar()
bar.add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
bar.add_yaxis('商家', [5, 20, 40, 10, 70, 90])

# render: 渲染,会渲染成html文件，默认的文件名是render.html
# bar.render()
bar.render('bar.html')
```

```python
# 在notebook中渲染
bar.render_notebook()
```

pyecharts 所有方法均支持链式调用。

```python
bar = (
    Bar()
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家', [5, 20, 40, 10, 70, 90])
)
bar.render_notebook()
```

#### 使用 options 配置项
- 在 pyecharts 中，一切皆 Options。

```python
from pyecharts import options as opts
```

```python
bar = (
    Bar()
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家', [5, 20, 40, 10, 70, 90])
    
    # 配置项
    # .set_global_opts(title_opts=opts.TitleOpts(title='主标题', subtitle='副标题'))
    # 或
    .set_global_opts(title_opts={'text': '主标题', 'subtext': '副标题'})
)
bar.render_notebook()
```

```python
bar = Bar()
    
bar.add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
bar.add_yaxis('商家', [5, 20, 40, 10, 70, 90])
    
bar.set_global_opts(title_opts=opts.TitleOpts(title='主标题', subtitle='副标题'))

bar.render_notebook()
```

使用主题

```python
# 导入主题包
from pyecharts.globals import ThemeType

bar = (
    Bar(
        init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    )
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家A', [5, 20, 40, 10, 70, 90])
    .add_yaxis('商家B', [15, 2, 50, 20, 40, 60])

)

bar.render_notebook()
```

在使用 Pandas和Numpy 时，请确保将数值类型转换为 python 原生的 int/float。比如整数类型请确保为 int，而不是 numpy.int32
